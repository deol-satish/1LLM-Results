```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
sns.set_style("darkgrid")

# Data for trainable parameter percentages
methods = ['Full Fine-Tune', 'NetLLM', 'L4S-LLM']
percentages = [100, 0.31, 1.005]

# Data for GPU memory consumption (in GB)
gpu_memory = [65.88, 27.24, 23.838]

# Plot for trainable parameter percentages
plt.figure(figsize=(6, 4), dpi=500)
plt.bar(methods, percentages, color=['blue', 'green', 'orange'])
plt.ylabel('Trainable Parameters (%)-Fraction of Lama2-7B Parameters')

# Manually annotate each bar with the percentage value
plt.text(0, percentages[0] + 1, f'{percentages[0]:.2f}%', ha='center', va='bottom')
plt.text(1, percentages[1] + 1, f'{percentages[1]:.2f}%', ha='center', va='bottom')
plt.text(2, percentages[2] + 1, f'{percentages[2]:.2f}%', ha='center', va='bottom')

plt.grid(True)

# Show the first plot
plt.tight_layout()
plt.savefig("Fratcion_Traininable_Parameters_Graph.png", dpi=500, bbox_inches='tight')
plt.savefig("Fratcion_Traininable_Parameters_Graph.eps", dpi=500, bbox_inches='tight')
plt.savefig("Fratcion_Traininable_Parameters_Graph.pdf", dpi=500, bbox_inches='tight')
plt.show()

# Plot for GPU memory consumption
plt.figure(figsize=(6, 4), dpi=500)
plt.bar(methods, gpu_memory, color=['blue', 'green', 'orange'])
plt.ylabel('GPU Memory Utilization (GB)')

# Manually annotate each bar with the memory value
plt.text(0, gpu_memory[0] + 1, f'{gpu_memory[0]:.2f} GB', ha='center', va='bottom')
plt.text(1, gpu_memory[1] + 1, f'{gpu_memory[1]:.2f} GB', ha='center', va='bottom')
plt.text(2, gpu_memory[2] + 1, f'{gpu_memory[2]:.2f} GB', ha='center', va='bottom')


plt.grid(True)

# Show the second plot
plt.tight_layout()


plt.savefig("LLM_GPU_Mem_Util_Graph.png", dpi=500, bbox_inches='tight')
plt.savefig("LLM_GPU_Mem_Util_Graph.eps", dpi=500, bbox_inches='tight')
plt.savefig("LLM_GPU_Mem_Util_Graph.pdf", dpi=500, bbox_inches='tight')


plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set plotting style
sns.set_style("darkgrid")

# Data for the average time to generate an answer
models = ['Llama2-7B', 'OPT-1.3B', 'NetLLM']
times = [0.2, 0.04, 0.03124]  # in seconds

# Create a bar plot for average time to generate an answer
plt.figure(figsize=(6, 4), dpi=500)
bar_width = 0.25  # Adjusted bar width for closer spacing
bar_positions = np.arange(len(models))

# Create the bars
plt.bar(bar_positions, times, width=bar_width, color=['blue', 'green', 'orange'])

# Add labels and title
plt.xlabel('Models')
plt.ylabel('Average Answer Generation Time (s)')
# plt.title('Average Time to Generate Answer for Different Models')  # Optional title

# Custom x-ticks for model names
plt.xticks(bar_positions, models)

# Adding value labels on top of each bar
for i, time in enumerate(times):
    plt.text(i, time , f'{time:.3f}s', ha='center')

# Grid settings
plt.grid(True)

# Show the plot
plt.tight_layout()

plt.savefig("LLM_Ans_Gen_Time_Graph.png", dpi=500, bbox_inches='tight')
plt.savefig("LLM_Ans_Gen_Time_Graph.eps", dpi=500, bbox_inches='tight')
plt.savefig("LLM_Ans_Gen_Time_Graph.pdf", dpi=500, bbox_inches='tight')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set plotting style
sns.set_style("darkgrid")

# Data
models = ['Llama2-7B', 'NetLLM']
valid_answer_percentages = [98, 99.96]

# Create custom positions for the bars to reduce distance
positions = np.array([0, 0.1])  # Adjusted positions for closer bars

# Create a bar plot
plt.figure(figsize=(6, 4), dpi=500)
bar_width = 0.025  # Width of the bars
bars = plt.bar(positions, valid_answer_percentages, color=['blue', 'orange'], width=bar_width)

# Adding data labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f"{yval}%", ha='center', va='bottom')

# Set x-ticks to model names, centered under each bar
plt.xticks(positions, models)  # Center the ticks under the bars

# Set labels
plt.ylabel('Generating Valid Answer (%)')

# Remove top and right spines completely
ax = plt.gca()  # Get the current Axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set y-axis limit for better visualization
plt.ylim(0, 110)

# Grid settings
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
